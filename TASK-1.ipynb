```python
df = df.dropna()
```

```python
df = df.dropna(subset=['CustomerID'])
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df.reset_index(drop=True, inplace=True)
```

```python
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

```python

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['CountryEncoded'] = le.fit_transform(df['Country'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Quantity_scaled', 'UnitPrice_scaled', 'TotalAmount_scaled']] = scaler.fit_transform(
    df[['Quantity', 'UnitPrice', 'TotalAmount']]
)
```

```python
customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',            
    'Quantity': 'sum',                 
    'TotalAmount': 'sum',             
    'InvoiceDate': ['min', 'max']     
}).reset_index()
customer_df.columns = ['CustomerID', 'InvoiceCount', 'TotalQuantity', 'TotalSpend', 'FirstPurchase', 'LastPurchase']
latest_date = df['InvoiceDate'].max()
customer_df['Recency'] = (latest_date - customer_df['LastPurchase']).dt.days
```

```python
rfm_df = customer_df[['CustomerID', 'Recency', 'InvoiceCount', 'TotalSpend']]
rfm_df.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
df.to_csv("Cleaned_Online_Retail.csv", index=False)
rfm_df.to_csv("Customer_Segmentation.csv", index=False)
```
